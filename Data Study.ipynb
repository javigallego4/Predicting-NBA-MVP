```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import colors
import seaborn as sns
import warnings
```

```python
df_stats = pd.read_csv("Data/player_mvp_stats.csv")
```

```python
correlacion_share_serie = df_stats.corr()['Share']
correlacion = pd.DataFrame(correlacion_share_serie)
```

```python
# Como el DataFrame tiene 38 filas vamos a dividirlo en dos df de 19 cada una para mejorar la interpretación 
# de la futura gráfica
correlacion1 = pd.DataFrame(correlacion[:19])
correlacion1.loc['Share']=[1]
correlacion2 = correlacion[-19:]
```

```python
f,ax = plt.subplots(figsize=(19,1.5),facecolor='white')
sns.color_palette("rocket", as_cmap=True)          # Esta paleta es la que viene por defecto
sns.heatmap(correlacion1.transpose(),annot = True,square=True, linewidths=1.5, cmap='rocket')
```

```python
f,ax = plt.subplots(figsize=(18,1.5))
sns.heatmap(correlacion2.transpose(), annot = True,square=True, linewidths=1.5, cmap='rocket')
```

```python
votos = df_stats[df_stats['Pts Won']>0]
```

```python
# Veamos una gráfica ilustrativa comparando los votos recibidos en función de puntos, rebotes y asistencias
sns.set(rc={'axes.facecolor':'lightgrey'})
sns.color_palette("rocket", as_cmap=True)
sns.pairplot(votos[['PTS','AST','TRB','Pts Won','Pos']],height=3,diag_kind="kde",hue='Pos',palette='rocket',corner=True)
```

```python
# Ahora realizaremos una gráfica comparativa entre los votos recibidos y el número de victorias y derrotas del
# equipo de cada jugador en una temporada concreta
sns.pairplot(votos[['W/L%','Pts Won','Pos']],height=3,diag_kind="kde",hue='Pos',palette='rocket',corner=True)
```

```python
#gráfica ilustrativa comparando los votos recibidos en función porcentaje de acierto en triples, tiros de 2 y 
# tiros libres
sns.pairplot(votos[['3P%','2P%','FT%','Pts Won','Pos']],height=3, diag_kind="kde",hue='Pos',palette='rocket',corner=True)
```
