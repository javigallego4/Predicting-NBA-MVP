```python
years = list(range(1991,2022))
```

```python
url_start = "https://www.basketball-reference.com/awards/awards_{}.html"
```

```python
import requests

for year in years: 
    url = url_start.format(year)              # Esto sustituye los '{}' por el año correspondiente
    data = requests.get(url)                  # Con esto descargamos la página web
    
    with open("MVPs/{}.html".format(year), "w+") as f:        # Con estas dos líneas de código estamos creando los
        f.write(data.text)                                    # archivos .html de cada uno de los años
```

```python
# Inspeccionamos la tabla de la que queremos scrapear los datos, para ver su 'nombre' en el .html
!pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup
```

```python
with open("MVPs/1991.html") as f:
    page = f.read()

#page 
soup = BeautifulSoup(page, "html.parser")
```

```python
# La cabecera de la tabla no nos sirve, entonces la borramos primero
soup.find('tr', class_="over_header").decompose()
# Como solo nos interesa una tabla vamos a borrar el resto. Esto lo haremos encontrando la tabla que queremos en
# concreto. Para ello, buscamos su 'id', que es único en todo el archivo .html. Entonces, vemos que este es "mvp".
mvp_table = soup.find(id="mvp")
```

```python
# Pandas es capaz de leer tablas de tipo html
import pandas as pd
!pip install lxml
mvp_1991 = pd.read_html(str(mvp_table))             # El argumento tiene que ser un string
```

```python
#mvp_1991                  # Vemos entonces que esto no es un DataFrame, sino una lista de ellos. Entonces: 
mvp_1991 = pd.read_html(str(mvp_table))[0]             # El argumento tiene que ser un string
mvp_1991
```

```python
dfs = []

for year in years:
    with open("MVPs/{}.html".format(year)) as f:
        page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_="over_header").decompose()
    mvp_table = soup.find(id="mvp")
    mvp = pd.read_html(str(mvp_table))[0]             # El argumento tiene que ser un string
    mvp['Year'] = year                                # Como no hay forma de ver a que año pertenecen los datos creamos una columna 'Year'
        
    
    dfs.append(mvp)
```

```python
# dfs           # Tenemos una lista de DataFrames para cada uno de los años
```

```python
# Obviamente, como no queremos trabajar con una lista de DataFrames, vamos a combinarlos. 
mvps = pd.concat(dfs)
# Hemos conseguido entonces un único DataFrame con todos los datos de los MVPs entre 1991-2021
mvps.head()
#mvps.tail()
```

```python
mvps.to_csv("mvps.csv")         # Guardamos el DataFrame en un archivo para poder trabajar con él fácilmente
```

```python
# Si queremos predecir quién va a ser el MVP no solo podemos tomar las votaciones de cada año. Tenemos que tomar
# estadísticas de cada uno de los jugadores que hay en la NBA. Entonces vamos a scrapear todas las estadísticas de 
# jugadores desde 1991-2021
```

```python
player_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html"

# Vamos a descargar primero los datos de un año concreto. 
url = player_stats_url.format(1991)
data = requests.get(url)
with open("Players/1991.html", "w+") as f: 
    f.write(data.text)
    
# Si comprobamos el contenido del archivo .html que acabamos de crear vemos que solo tiene 17 filas. Sin embargo,
# en el navegador hay muchas más. La razón de esto es que esto usa JavaScrypt. Entonces usaremos Selenium para
# scrapear páginas hechas con JavaScrypt
```
